```python
import tkinter as tk
from tkinter import scrolledtext

# Predefined responses
responses = {
    "hello": "Hello! I am Bot. How can I assist you today?",
    "hi": "Hi there! I am Bot. How can I help?",
    "how are you": "I'm just a bot, but I'm doing great! How about you?",
    "price": "Our products range from ₹500 to ₹5000. What are you looking for?",
    "refund": "Refunds are processed within 5-7 business days.",
    "support": "You can contact our support team at support@bot.com.",
    "bye": "Goodbye! Have a great day!",
    "exit": "Goodbye! Have a great day!",
    "discount": "We currently offer a 20% discount on all items!",
    "shipping": "Standard shipping takes 3-5 business days. Express options are available too!"
}

def get_response(user_input):
    user_input = user_input.lower()
    if "hello" in user_input:
        return responses["hello"]
    elif "hi" in user_input:
        return responses["hi"]
    elif "how are you" in user_input:
        return responses["how are you"]
    elif "price" in user_input:
        return responses["price"]
    elif "refund" in user_input:
        return responses["refund"]
    elif "support" in user_input:
        return responses["support"]
    elif "bye" in user_input or "exit" in user_input:
        return responses["bye"]
    elif "discount" in user_input:
        return responses["discount"]
    elif "shipping" in user_input:
        return responses["shipping"]
    else:
        return "I'm sorry, I don't understand. Can you rephrase?"

def send_message():
    """Handles sending user messages and displaying responses."""
    user_message = user_entry.get().strip()
    if user_message == "":
        return
    chat_window.config(state=tk.NORMAL)
    chat_window.insert(tk.END, "You: " + user_message + "\n", "user")
    user_entry.delete(0, tk.END)

    bot_response = get_response(user_message)
    chat_window.insert(tk.END, "Bot: " + bot_response + "\n\n", "bot")

    chat_window.config(state=tk.DISABLED)
    chat_window.yview(tk.END)

# GUI setup
root = tk.Tk()
root.title("Customer Support Chatbot")
root.geometry("420x500")
root.resizable(False, False)
root.configure(bg="#F0F0F0")

# Chat window
chat_window = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=50, height=20,
                                        state=tk.DISABLED, bg="#FFFFFF", fg="black", font=("Arial", 12))
chat_window.pack(pady=10, padx=10)

# Apply tag styles
chat_window.tag_config("user", foreground="#0084FF", font=("Arial", 12, "bold"))
chat_window.tag_config("bot", foreground="#28A745", font=("Arial", 12, "bold"))

# Entry field
user_entry = tk.Entry(root, width=40, font=("Arial", 12), bg="#FFFFFF", fg="black",
                      relief=tk.GROOVE)
user_entry.pack(pady=5)

# Button Styles
button_frame = tk.Frame(root, bg="#F0F0F0")
button_frame.pack(pady=5)

send_button = tk.Button(button_frame, text="Send", command=send_message,
                        font=("Arial", 12, "bold"), bg="#0084FF", fg="white",
                        relief=tk.RAISED, padx=10, pady=5)
send_button.pack(side=tk.LEFT, padx=10)

# Exit button with hover effects
def on_exit_hover(event):
    exit_button.config(bg="#D32F2F")

def on_exit_leave(event):
    exit_button.config(bg="#DC3545")

exit_button = tk.Button(button_frame, text="Exit", command=root.quit,
                        font=("Arial", 12, "bold"), bg="#DC3545", fg="white",
                        relief=tk.RAISED, padx=10, pady=5)
exit_button.pack(side=tk.RIGHT, padx=10)
exit_button.bind("<Enter>", on_exit_hover)
exit_button.bind("<Leave>", on_exit_leave)

# Run the chatbot
root.mainloop()
```
